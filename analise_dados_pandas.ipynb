Análise de Dados com Pandas

```python
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataset_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/refs/heads/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

urlretrieve(dataset_url, 'datasets/global_cases_covid19.csv')
```

```python
df_covid = pd.read_csv('datasets/global_cases_covid19.csv')
```

```python
df_covid
```

```python
df_covid.head()
```

```python
df_covid.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
df_covid
```

```python
df_country = df_covid.groupby('Country/Region').sum()
```

```python
df_country
```

```python
df_country.loc['Afghanistan']
```

```python
data = df_country.loc['Brazil'].index
cases  = df_country.loc['Brazil'].values
```

```python
plt.figure(figsize=(19,8))
plt.bar(data, cases)
```

```python
df_country.head()
```

```python
s_brazil = df_country.loc['Brazil']
```

```python
s_brazil = s_brazil[s_brazil > 0]
```

```python
tam = len(s_brazil)

plt.figure(figsize=(29,8))
plt.xticks(rotation = 60)
plt.bar(s_brazil.index[tam-200:tam], s_brazil.values[tam-200:tam])
```
